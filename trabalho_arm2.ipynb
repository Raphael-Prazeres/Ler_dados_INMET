```python
def leia_arquivos_xlsx():
    # Cria uma lista com os nomes dos arquivos
    arquivos = [f"Dados/31787_{ano}.xlsx" for ano in range(12, 23)]

    # Lê cada arquivo e adiciona ao dataframe
    dfs = []
    for arquivo in arquivos:
        dfs.append(pd.read_excel(arquivo))

    # Concatena os dataframes
    df = pd.concat(dfs)

    # Retorna o dataframe
    return df


# Imprime o dataframe
#print(leia_arquivos_xlsx())
```
